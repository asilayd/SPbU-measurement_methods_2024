### Обработка результатов

1. Вычисление массы используемых в опыте шариков по формуле: m=V*ρ, где ρ – плотность вещества, из которого изготовлен шарик, V – объем шарика (размеры и плотность взяты из указаний в пособии). <br/>
Будем считать объем по известной формуле объема шара: <br/>
$V_{шара} = \frac {4}{3} \pi R^3$ при R = 5 мм. <br/>
2. Нахождение среднего значения времени полета шариков по формуле: <br/>
$\overline t = \frac {\sum \limits_{i=1}^n t_i} {n} $

```python
#table1.columns[0]
#table1[table1.columns[0]].to_list()[1:]
names = ["Алюминий", "Латунь", "Сталь", "Дерево", "Плексиглас", "Свинец"]
time_data = {}
for n in names:
    time_data[n] = list(map(float, table1[n].to_list()[1:]))
average_val = [sum(time_data[key])/len(time_data[key]) for key in time_data]

average_data = {"Вещество": ["t м/c"], }
i  =0
for n in names:
    average_data[n] = average_val[i]
    i+=1

table_average = pd.DataFrame(average_data)
table_average.to_csv("table_a.scv", index = False)
table3 = pd.read_csv("table_a.scv", index_col=0)
print("Получим средние значения")
table3
```

3. Найдем стандартную погрешность измерений по формуле $\Delta t = \sqrt {\frac {\sum \limits_{i=1}^n d_i^2} {n(n-1)} }, где \: d_i = t_i - \overline t $

```python
N = 30
dev_data = {"Вещество": ["$ \Delta t мc$"], }
for n in names:
    dev_data[n] = ((sum([(t-average_data[n])**2 for t in time_data[n]]))/(30*(30-1)))**0.5

table_dev = pd.DataFrame(dev_data)
table_dev.to_csv("table_d.scv", index = False)
table4 = pd.read_csv("table_d.scv", index_col=0)
print("Стандартная погрешность времени падения")
table4
```

4. Вычислим ускорение свободного падения g по формуле $g = \frac {2(h-v_0 t)} {t^2} $ , где $ h = 0.272 \pm 0.001 м, v_0 = 1.050 \pm 0.005 м/c $

```python
h = 0.272
v = 1.050
g_list = list()
for n in names:
    g_list.append((2*(h-v*average_data[n]/1000))/(average_data[n]/1000)**2)
g_data = {"№ п/п": range(1, 7), "Вещество": names, "g, м/c": g_list}
g_table = pd.DataFrame(g_data)
g_table.to_csv("table5.csv", index = False)
table5 = pd.read_csv("table5.csv", index_col = 0)
print("Вычисление ускорения свободного падения ")
table5
```

Для погрешности измерений пользуемся формулой $\Delta g = \sqrt { (\frac {-4h} {t^3} + \frac {v_0} {t^2} )^2 (\Delta t)^2 + (\frac {2}{3t^2})^2 (\Delta h)^2 + (\frac {-2}{3t})^2(\Delta v_0)^2 } $

```python
d_h = 0.0011
d_v = 0.0051 
delta_g_list = list()
def g(h, t, v, d_t, d_h, d_v):
    return (( (-4*h/(t**3) + v/(t**2))**2 )*d_t**2 + (2/(3*t**2))**2 * (d_h)**2 + (-2/(3*t))**2 * (d_v)**2)**0.5
for n in names:
    delta_g_list.append(g(h, average_data[n], v, dev_data[n], d_h, d_v ))
delta_g_data = {"№ п/п": range(1,7),"Вещество": names, "$\Delta g \: м/c^2$" : delta_g_list}
delta_g_table = pd.DataFrame(delta_g_data)
delta_g_table.to_csv("table6.csv", index=False)
table6 = pd.read_csv("table6.csv", index_col = 0)
print("Стандартная погрешность ускорения свободного падения")
table6
```
